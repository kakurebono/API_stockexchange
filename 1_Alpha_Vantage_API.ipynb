```python
def return_inv(col):
    
    return ((col-col.shift(-1))/col.shift(-1)) * 100 #formula for calculated stock return 
```

```python
stock_data = stock_data.astype(float) # change data type to float
```

```python

stock_data[['AAPL_returns', 'MSFT_returns', 'AMZN_returns', 'TSLA_returns']] = stock_data[['AAPL', 'MSFT', 'AMZN', 'TSLA']].apply(return_inv)
#adding new rows with the calculated return price fileds 
```

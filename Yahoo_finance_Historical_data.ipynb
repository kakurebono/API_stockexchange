```python
stock_url = 'https://query1.finance.yahoo.com/v7/finance/download/{}?'

# params = {
#     'period1':'1575909637',
#     'period2':'1607532037',
#     'interval' : '1d',
#     'events' : 'history',
#     'includeAdjustedClose' : 'true'
# }
```

```python
params = {
    'range' : '1y',
    'interval' : '1d',
    'events' : 'history'   
}
```

```python
response = requests.get(stock_url.format(stock), params=params) # url reader
```

```python
file = StringIO(response.text) # reads the data 
reader = csv.reader(file)
data = list(reader)
```

```python
df = pd.DataFrame(data) #dataframe from the pulled data
df
```

```python
df = df[[0,5]] # keep date and Adj close only
df = df.iloc[1:] # pop first row
```

```python
df.rename(columns = {'0':'Date', '5' : 'stock' }, inplace=True)
# df.add_prefix('X_')
```
